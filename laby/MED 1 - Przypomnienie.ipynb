```python
from IPython.display import display

# wybór cech numerycznych
num_d = d.select_dtypes(include=[np.number]).columns.tolist()
num_iris = iris.select_dtypes(include=[np.number]).columns.tolist()

print('\n=== Zbiór d — podstawowe miary (cały zbiór) ===')
if len(num_d) > 0:
    display(d[num_d].agg(['mean','median','std']).T)
else:
    print('Brak cech numerycznych w zbiorze d')

print('\n=== Zbiór d — miary wg klas ===')
if 'klasa' in d.columns and len(num_d) > 0:
    display(d.groupby('klasa')[num_d].agg(['mean','median','std']).T)
else:
    print('Brak kolumny "klasa" lub brak cech numerycznych w d')

print('\n=== Zbiór iris — podstawowe miary (cały zbiór) ===')
if len(num_iris) > 0:
    display(iris[num_iris].agg(['mean','median','std']).T)
else:
    print('Brak cech numerycznych w iris')

print('\n=== Zbiór iris — miary wg gatunku ===')
if 'species' in iris.columns and len(num_iris) > 0:
    display(iris.groupby('species')[num_iris].agg(['mean','median','std']).T)
else:
    print('Brak kolumny "species" lub brak cech numerycznych w iris')

# Wykresy: histogram + boxplot dla każdej cechy w d
for col in num_d:
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    sns.histplot(data=d, x=col, kde=True)
    plt.title(f'Histogram (d): {col}')
    plt.subplot(1,2,2)
    if 'klasa' in d.columns:
        sns.boxplot(x='klasa', y=col, data=d)
        plt.title(f'Boxplot (d) — {col} wg klasy')
    else:
        sns.boxplot(y=col, data=d)
        plt.title(f'Boxplot (d) — {col}')
    plt.tight_layout()
    plt.show()

# Macierz punktowa (pairplot) dla d (jeżeli jest sens)
try:
    if len(num_d) >= 2:
        if 'klasa' in d.columns:
            sns.pairplot(d[num_d + ['klasa']], hue='klasa')
        else:
            sns.pairplot(d[num_d])
        plt.show()
except Exception as e:
    print('pairplot(d) nie powiódł się:', e)

# Wykresy dla iris
for col in num_iris:
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    sns.histplot(data=iris, x=col, kde=True)
    plt.title(f'Histogram (iris): {col}')
    plt.subplot(1,2,2)
    if 'species' in iris.columns:
        sns.boxplot(x='species', y=col, data=iris)
        plt.title(f'Boxplot (iris) — {col} wg species')
    else:
        sns.boxplot(y=col, data=iris)
        plt.title(f'Boxplot (iris) — {col}')
    plt.tight_layout()
    plt.show()

# Pairplot dla iris
try:
    if len(num_iris) >= 2 and 'species' in iris.columns:
        sns.pairplot(iris, hue='species')
        plt.show()
except Exception as e:
    print('pairplot(iris) nie powiódł się:', e)

```

> **Zadanie 4** Zastanów się, jakie wnioski odnoszące się do dystrybucji danych można wyciągnąć, analizując jednocześnie dwie miary tendencji centralnej - średnią i medianę. Wykonaj analizę wartości obu tych miar dla zbiorów `iris` oraz `panstwa`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

def analyze_mean_median(df, group_col=None, title_prefix=''):
    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    if not num_cols:
        print(title_prefix, "Brak cech numerycznych")
        return

    # Tabela mean / median i różnica
    stats = df[num_cols].agg(['mean','median']).T
    stats['diff'] = stats['mean'] - stats['median']
    stats['rel_diff_%'] = 100 * stats['diff'] / stats['median'].replace(0, np.nan)
    display(stats.round(4))

    # Miary wg grup jeśli podano
    if group_col and group_col in df.columns:
        print(f"\n{title_prefix} — mean/median wg {group_col}:")
        display(df.groupby(group_col)[num_cols].agg(['mean','median']).round(4))

    # Wykresy: histogram z liniami mean/median (bez dodatkowych miar)
    for col in num_cols:
        plt.figure(figsize=(8,3))
        sns.histplot(df[col], kde=True, color='C0', alpha=0.6)
        plt.axvline(df[col].mean(), color='red', linestyle='--', label='mean')
        plt.axvline(df[col].median(), color='green', linestyle=':', label='median')
        plt.title(f"{title_prefix} — {col}: mean vs median")
        plt.legend()
        plt.show()

        # krótka interpretacja automatyczna
        m = df[col].mean()
        med = df[col].median()
        rel = 100 * (m - med) / (abs(med) if med != 0 else 1)
        if abs(rel) < 5:
            note = "mean ≈ median więc rozkład bliski symetrii / brak silnych wartości odstających"
        elif rel > 0:
            note = "mean > median więc największy wpływ mają duże wartości"
        else:
            note = "mean < median więc największy wpływ mają małe wartości"
        print(f"{col}: mean={m:.3f}, median={med:.3f}, rel_diff={rel:.1f}% -> {note}\n")

# Uruchomienie analizy tylko dla mean i median:
analyze_mean_median(iris, group_col='species', title_prefix='IRIS')
num_panstwa = panstwa.select_dtypes(include=[np.number]).columns.tolist()
print("\nNumeryczne kolumny w panstwa:", num_panstwa)
analyze_mean_median(panstwa, title_prefix='PANSTWA')
```

**Odpowiedź** 
Jeżeli średnia jest większa od mediany to oznacza, że wpływ dużych wartości na średnią jest duży. W przeciwnym przypadku - wpływ małych wartości jest duży na średnią. Gdy średnia i mediana są zbliżone - oznacza to, że wartości odstających nie ma lub nie ma mają na tyle duzego wplływu aby zaburzyć średnią.

Miary współzależności pozwalają na określenie czy i w jakim stopniu poszczególne atrybuty są od siebie zależne. Podstawową miarą zależności liniowej jest korelacja.
